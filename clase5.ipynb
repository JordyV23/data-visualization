```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

```python
# ruta del archivo de datos
iris_filepath = "./data/iris.csv"

# lee el archivo de datos en un DataFrame
iris_data = pd.read_csv(iris_filepath,index_col="Id")

iris_data.head()
```

```python
# histograma de los datos
sns.histplot(iris_data['Petal Length (cm)'])
```

```python
# Kenerl Density Estimate(KDE) plot
sns.kdeplot(data=iris_data['Petal Length (cm)'], fill=True)
```

```python
# 2D KDE plot
sns.jointplot(x=iris_data['Petal Length (cm)'], y=iris_data['Sepal Width (cm)'], kind="kde")
```

```python
# Histogramas para cada especie
sns.histplot(data=iris_data, x ='Petal Length (cm)', hue='Species')

# Agrear titulo
plt.title("Histograma de longitud de petalo, por especies")
```

```python
# KDE plot para cada especie
sns.kdeplot(data=iris_data, x ='Petal Length (cm)', hue='Species',shade=True)

# Agrear titulo
plt.title("Distribucion de longitud de petalo, por especies")
```
